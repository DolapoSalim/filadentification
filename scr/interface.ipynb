```python
import gradio as gr
import tensorflow as tf
import numpy as np
from PIL import Image
```

```python

model_path = r'C:\Users\dolap\OneDrive\Documents\DOLAPO\data-analysis\filadentification\model\filadentification.h5'

# Load your trained model
model = tf.keras.models.load_model(model_path)

# Define class names
class_names = ["fila", "not-fila"]  

def predict(image):
    # Preprocess the image
    image = image.resize((256, 256))  # Resize to model input size
    image = np.array(image) / 255.0   # Normalize
    image = np.expand_dims(image, axis=0)  # Add batch dimension

    # Predict
    prediction = model.predict(image)

    # If model outputs a single value (sigmoid) or two values (softmax)
    if prediction.shape[-1] == 1:
        prob = prediction[0][0]
        predicted_class = class_names[int(prob > 0.5)]
        confidence = prob if prob > 0.5 else 1 - prob
    else:
        prob = tf.nn.softmax(prediction[0]).numpy()
        predicted_class = class_names[np.argmax(prob)]
        confidence = np.max(prob)

    confidence_percentage = confidence * 100
    result_text = f"Prediction: image contains **{predicted_class}**\nConfidence: **{confidence_percentage:.2f}%**"

    return result_text

# Create Gradio interface
interface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil", label="Upload an image"),
    outputs=gr.Markdown(),
    title="Fila Yoruba Detector",
    description="Upload an image. This model will predict whether the image contains the traditional Yoruba people's **fila** or not",
    theme="default"  # You can also try "huggingface" or "soft" themes
)

# Launch the app
interface.launch(share=True)  

```
